```python
import numpy as np
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('GN.csv')
```

```python
data
```

```python
data.head()
```

```python
X = data['credit_amount'].values
Y = data['duration'].values
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,Y)
```

```python
m = len(X)

X = X.reshape((m, 1))
```

```python
reg = LinearRegression()

reg = reg.fit(X, Y)
```

```python
Y_pred = reg.predict(X)
```

```python
plt.plot(X,Y_pred)
```

```python
plt.plot(X,Y_pred)
plt.scatter(X,Y)
plt.show()
```
